```python
!pip3 install instaloader

import instaloader
import re

def get_instagram_hashtags(username):
    L = instaloader.Instaloader()
    
    # Laden des Instagram-Profils
    profile = instaloader.Profile.from_username(L.context, username)
    
    hashtags = []

    # Durchsuchen der letzten Posts (10 st., nach Bedarf ab√§ndern)
    for post in profile.get_posts():
        caption = post.caption
        if caption:
            post_hashtags = re.findall(r'#\w+', caption)
            hashtags.extend(post_hashtags)
        
        if len(hashtags) >= 10:  # Limit auf 10 Posts
            break

    return hashtags

# Benutzername des Influencers festlegen
username = 'alexa_breit'  
hashtags = get_instagram_hashtags(username)

print(hashtags)
```

```python
import pandas as pd

# Instaloader-Instanz erstellen
L = instaloader.Instaloader()

# Profilname definieren
profile_name = 'alexa_breit'

# Profil laden
profile = instaloader.Profile.from_username(L.context, profile_name)

# Datenstruktur vorbereiten
data = []

# Posts durchlaufen und Daten extrahieren
for post in profile.get_posts():
    caption = post.caption
    hashtags = post.caption_hashtags
    data.append({'profile': profile_name, 'caption': caption, 'hashtags': hashtags})

# DataFrame erstellen
df = pd.DataFrame(data)

# Daten anzeigen
df
```
